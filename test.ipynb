```python
import numpy as np
import matplotlib.pyplot as plt
import json
import importlib
import trajectory_points
importlib.reload(trajectory_points)

from tools import VARIABLES, saved_data_update


var = VARIABLES()
saved_data_update(var)



def getPoints(str):
    pnts = str.split(' ')
    return [float(i) for i in pnts]



def plot_init_sw(p):
    p1_init_sw_midLow = getPoints(p.init_sw_midLow)[:3]
    p1_init_sw_mid = getPoints(p.init_sw_mid)[:3]
    p1_init_sw_midUp = getPoints(p.init_sw_midUp)[:3]
    p1_sw_end = getPoints(p.sw_end)[:3]
    
    # PLOT INIT SWING XZ ============================================================
    plt.figure(figsize=(10,10))
    plt.title(' INIT SWING XZ')
    plt.xlabel('X')
    plt.ylabel('Z')

    plt.scatter(p1_init_sw_midLow[0], p1_init_sw_midLow[2])     
    plt.scatter(p1_init_sw_mid[0], p1_init_sw_mid[2])           
    plt.scatter(p1_init_sw_midUp[0], p1_init_sw_midUp[2])       
    plt.scatter(p1_sw_end[0], p1_sw_end[2])    
    plt.scatter(p.initP.x, p.initP.z, marker='x')                 
    plt.legend(['init_sw_midLow', 'init_sw_mid', 'init_sw_midUp', 'swing_end', 'init'])

    # PLOT INIT SWING YZ ============================================================
    plt.figure(figsize=(10,10))
    plt.title(' INIT SWING YZ')
    plt.xlabel('Y')
    plt.ylabel('Z')
    # XZ :
    plt.scatter(p1_init_sw_midLow[1], p1_init_sw_midLow[2])     # p1_1
    plt.scatter(p1_init_sw_mid[1], p1_init_sw_mid[2])           # p1_2
    plt.scatter(p1_init_sw_midUp[1], p1_init_sw_midUp[2])       # p1_3
    plt.scatter(p1_sw_end[1], p1_sw_end[2])                     # p1_5
    plt.scatter(p.initP.y, p.initP.z, marker='x')
    plt.legend(['init_sw_midLow', 'init_sw_mid', 'init_sw_midUp', 'swing_end', 'init'])



def plot_main_sw(p):
    p1_sw_start = getPoints(p.st_end)[:3]
    p1_sw_midLow = getPoints(p.sw_midLow)[:3]
    p1_sw_mid = getPoints(p.sw_mid)[:3]
    p1_sw_midUp = getPoints(p.sw_midUp)[:3]
    p1_sw_end = getPoints(p.sw_end)[:3]

    # PLOT MAIN SWING XZ ============================================================
    plt.figure(figsize=(10,10))
    plt.title(' MAIN SWING XZ')
    plt.xlabel('X')
    plt.ylabel('Z')

    plt.scatter(p1_sw_start[0], p1_sw_start[2])
    plt.scatter(p1_sw_midLow[0], p1_sw_midLow[2])       # 
    plt.scatter(p1_sw_mid[0], p1_sw_mid[2])
    plt.scatter(p1_sw_midUp[0], p1_sw_midUp[2])
    plt.scatter(p1_sw_end[0], p1_sw_end[2])
    plt.scatter(p.initP.x, p.initP.z, marker='x')
    plt.legend(['sw_start', 'sw_midLow', 'sw_mid', 'sw_midUp', 'sw_end', 'init'])

    # PLOT MAIN SWING YZ ============================================================
    plt.figure(figsize=(10,10))
    plt.title(' MAIN SWING YZ')
    plt.xlabel('Y')
    plt.ylabel('Z')

    plt.scatter(p1_sw_start[1], p1_sw_start[2])
    plt.scatter(p1_sw_midLow[1], p1_sw_midLow[2])       # 
    plt.scatter(p1_sw_mid[1], p1_sw_mid[2])
    plt.scatter(p1_sw_midUp[1], p1_sw_midUp[2])
    plt.scatter(p1_sw_end[1], p1_sw_end[2])
    plt.scatter(p.initP.y, p.initP.z , marker='x')
    plt.legend(['sw_start', 'sw_midLow', 'sw_mid', 'sw_midUp', 'sw_end', 'init'])


def plot_st(p):
    p1_st_start = getPoints(p.sw_end)[:3]
    p1_st_midLow = getPoints(p.st_midLow)[:3]
    p1_st_mid = getPoints(p.st_mid)[:3]
    p1_st_midUp = getPoints(p.st_midUp)[:3]
    p1_st_end = getPoints(p.st_end)[:3]

    # PLOT STANCE XZ ============================================================
    plt.figure(figsize=(10,10))
    plt.title(' MAIN STANCE XZ')
    plt.xlabel('X')
    plt.ylabel('Z')
    plt.scatter(p1_st_start[0],     p1_st_start[2])
    plt.scatter(p1_st_midUp[0],     p1_st_midUp[2])
    plt.scatter(p1_st_mid[0],       p1_st_mid[2])
    plt.scatter(p1_st_midLow[0],    p1_st_midLow[2])
    plt.scatter(p1_st_end[0],       p1_st_end[2])
    plt.scatter(p.initP.x,          p.initP.z, marker='x')
    plt.legend(['st_start',  'st_midUp', 'st_mid', 'st_midLow', 'st_end', 'init'])

    # PLOT STANCE YZ ============================================================
    plt.figure(figsize=(10,10))
    plt.title(' MAIN STANCE YZ')
    plt.xlabel('Y')
    plt.ylabel('Z')
    plt.scatter(p1_st_start[1],     p1_st_start[2])
    plt.scatter(p1_st_midUp[1],     p1_st_midUp[2])
    plt.scatter(p1_st_mid[1],       p1_st_mid[2])
    plt.scatter(p1_st_midLow[1],    p1_st_midLow[2])
    plt.scatter(p1_st_end[1],       p1_st_end[2])
    plt.scatter(p.initP.y,          p.initP.z, marker='x')
    plt.legend(['st_start',  'st_midUp', 'st_mid', 'st_midLow', 'st_end', 'init'])



```
